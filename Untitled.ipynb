```python
import numpy as np
import torch
import matplotlib.pyplot as plt
%matplotlib inline
import h5py
from models import VAE
```

```python
test = np.load('./data/250k.npz')['arr'].astype(np.float32)
test = torch.from_numpy(test)
```

```python
model = VAE()
model.load_state_dict(torch.load('./models/vae-1.pth'))
model.eval()
Z = model.encode(test)[0].detach().numpy()
plt.scatter(Z[:, 0], Z[:, 1], s=1)
```
